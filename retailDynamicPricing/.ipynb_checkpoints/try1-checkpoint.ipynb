```python
import numpy as np
import pandas as pd
from scipy import stats

def outliers(data):
    # calculate z-scores for each data point
    z_scores = np.abs((data - np.mean(data)) / np.std(data))

    # define a threshold for outlier detection
    threshold = 3

    # identify outliers
    outliers = np.where(z_scores > threshold)

    # print the outliers
    print("Outliers:", outliers[0])
    
    return outliers[0]

    # remove outliers
```

```python
newDf = []
for cmbs in tshirt.IS_COMBO.unique():
    xx = tshirt[['SELL_ID','PRICE','QUANTITY']][tshirt['IS_COMBO']==cmbs]
    xx.reset_index(drop=True, inplace=True)
    print(xx.shape)
    print(cmbs)
    ch = outliers(xx)
    we = xx.drop(ch.tolist(), axis=0)
    print(we.shape)
    newDf.append(we)
```

```python
tshirtNew = pd.concat(newDf, ignore_index=True)
tshirtNew.head()
```

```python
from tpot import TPOTRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline, make_union
from sklearn.svm import LinearSVR
from tpot.builtins import StackingEstimator, ZeroCount
from tpot.export_utils import set_param_recursive
```

```python
withOutCombo = tshirtNew[tshirtNew['IS_COMBO']==0]
X_train, X_test, y_train, y_test = train_test_split(withOutCombo[['PRICE']], withOutCombo[['QUANTITY']], test_size=0.3, random_state=42)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)

# define evaluation procedure
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
# define search
tpotModelwithOutCombo = TPOTRegressor(generations=5, population_size=50, scoring='neg_mean_absolute_error', cv=cv, verbosity=2, random_state=42, n_jobs=-1)
# perform the search
tpotModelwithOutCombo.fit(X_train, y_train)
```

```python
#tpotModelwithOutCombo.export(r'./models/tshirtModelWithOutCombo.py')

bestModel_ = []
for idx, (name, transform) in enumerate(tpotModelwithOutCombo.fitted_pipeline_.steps, start=1):
    bestModel = transform
    #print(f'{idx}. {transform}')
    bestModel_.append(bestModel)
bestModel1_ = tuple(bestModel_)
```

```python
# exported_pipeline = make_pipeline(
#         ZeroCount(),
#         StackingEstimator(estimator=GradientBoostingRegressor(alpha=0.9, learning_rate=0.1, 
#                                                               loss="huber", max_depth=8, max_features=0.6000000000000001, 
#                                                               min_samples_leaf=2, min_samples_split=10, n_estimators=100, 
#                                                               subsample=0.6500000000000001)),
#         LinearSVR(C=0.0001, dual=True, epsilon=0.1, loss="epsilon_insensitive", tol=0.1))
```

```python
exported_pipeline = make_pipeline(*bestModel1_)
print(exported_pipeline)
```

```python
# Fix random state for all the steps in exported pipeline
set_param_recursive(exported_pipeline.steps, 'random_state', 42)

exported_pipeline.fit(X_train, y_train)
```

```python
exported_pipeline.steps[-1][1].coef_ #pipeline.steps[-1][1].coef_
```

```python
pred_tpot_tr = exported_pipeline.predict(X_train)
pred_tpot_te = exported_pipeline.predict(X_test)

mse_tr = mean_squared_error(y_train, pred_tpot_tr)
mse_te = mean_squared_error(y_test, pred_tpot_te)
print('mse_tr',mse_tr, 'mse_te',mse_te)

mae_tr = mean_absolute_error(y_train, pred_tpot_tr)
mae_te = mean_absolute_error(y_test, pred_tpot_te)
print('mae_tr',mae_tr, 'mae_te',mae_te)

mape_tr = mean_absolute_percentage_error(y_train, pred_tpot_tr)
mape_te = mean_absolute_percentage_error(y_test, pred_tpot_te)
print('mape_tr',mape_tr, 'mape_te',mape_te)

r2score_tr = r2_score(y_train, pred_tpot_tr)
r2score_te = r2_score(y_test, pred_tpot_te)
print('r2score_tr',r2score_tr, 'r2score_te',r2score_te)
```
