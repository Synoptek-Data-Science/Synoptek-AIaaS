```python
# Load the data
sold = pd.read_csv(r'./inputs/SaleMetaData_.csv')
transactions = pd.read_csv(r'./inputs/transactionsN1_.csv')
date_info = pd.read_csv(r'./inputs/DateInfo.csv')
```

```python
display(sold.head())
print(sold.shape)
```

```python
ch1 = sorted(sold.SELL_ID.unique().tolist())
print(len(ch1))
print(ch1)
```

```python
transactions['CALENDAR_DATE'] = pd.to_datetime(transactions['CALENDAR_DATE'])
transactions_ = transactions[transactions['SELL_ID']!=2054]
display(transactions_.head())
print(transactions_.shape)
```

```python
print(len(transactions_.SELL_ID.unique().tolist()))
print(sorted(transactions_.SELL_ID.unique().tolist()))
```

```python
date_info.drop(['IS_SCHOOLBREAK', 'AVERAGE_TEMPERATURE', 'IS_OUTDOOR'], axis=1, inplace=True)
date_info['CALENDAR_DATE'] = pd.to_datetime(date_info['CALENDAR_DATE'])
date_info['HOLIDAY'] = date_info['HOLIDAY'].fillna("No Holiday")
date_info.head()
```

```python
data1 = pd.merge(sold, transactions, on =  'SELL_ID')

b = data1.groupby(['SELL_ID', 'ITEM_ID','ITEM_NAME', 'CALENDAR_DATE','PRICE','IS_COMBO']).QUANTITY.sum()

intermediate_data = b.reset_index()

print(intermediate_data.shape)
```

```python
intermediate_data.head()
```

```python
#intermediate_data.to_csv(r'./inputs/newDataN.csv', index=False)
```

```python
combined_data = pd.merge(intermediate_data, date_info, on = 'CALENDAR_DATE')
combined_data.head()
```

```python
#combined_data.to_csv(r'./inputs/newDataN1.csv', index=False)
```

```python
combined_data.shape
```

```python
bau_data = combined_data[(combined_data['HOLIDAY']=='No Holiday') & \
                         (combined_data['IS_WEEKEND']==0)]
bau_data.reset_index(drop=True, inplace=True)
```

```python
bau_data.head()
```

```python
bau_data.shape
```

```python
# Data exploration
for itm in bau_data.ITEM_NAME.unique():
    ch = bau_data[bau_data['ITEM_NAME']==itm]
    sns.scatterplot('PRICE', 'QUANTITY', data=ch, label=itm)
    plt.show()
```

```python
tshirt = bau_data[bau_data['ITEM_NAME']=='T-SHIRT']
tshirt.reset_index(drop=True, inplace=True)
display(tshirt.head())
#sns.scatterplot('PRICE', 'QUANTITY', data=tshirt, label='T-SHIRT')
sns.histplot(data=tshirt, x='PRICE', hue='IS_COMBO')
plt.show()
```

```python
sns.scatterplot(data=tshirt, x='PRICE', y='QUANTITY', hue='IS_COMBO')
plt.show()
```

```python
sns.regplot(data=tshirt, x='PRICE', y='QUANTITY')
plt.show()
```

```python
sns.boxplot(data=tshirt, y='PRICE', x='IS_COMBO')
plt.show()
```
