# lets create a function which can create the models for each item with different sell ids and record the elasticities

```python
elasticities = {}
```

```python
def create_model_and_find_elasticity(data):
    model = ols("QUANTITY ~ PRICE", data).fit()
    price_elasticity = model.params[1]
    print("Price elasticity of the product: " + str(price_elasticity))
    print(model.summary())
    fig = plt.figure(figsize=(12,8))
    fig = sm.graphics.plot_partregress_grid(model, fig=fig)
    return price_elasticity, model
```

# Find optimal price for maximum profit

Now, let's take coke (the sell_id was 2053 for the last coke data) and since we do not the buying price of coke, let''s assume it to be a little less than the minimum coke price in the dataset

```python
coke2053_data.PRICE.min()
```

```python
coke2053_data.PRICE.max()
```

Let's take 9 as the buying price of coke. We now want to be able to set the price of coke to get the maximum profit. PRICE is the selling price

```python
buying_price_coke = 9
```

$$coke data.PROFIT = (coke data.PRICE - buying price coke) * coke data.QUANTITY$$
Let's see the profit for various price points:

```python
start_price = 9.5
end_price = 20
```

```python
for keys in elasticities:
    print(keys)
```

```python
def find_optimal_price(data, model, buying_price, sku_name):
    start_price = data.PRICE.min() - 1
    end_price = data.PRICE.min() + 10
    test = pd.DataFrame(columns = ["PRICE", "QUANTITY"])
    test['PRICE'] = np.arange(start_price, end_price, 0.01)
    test['QUANTITY'] = model.predict(test['PRICE'])
    test['PROFIT'] = (test["PRICE"] - buying_price) * test["QUANTITY"]
    
    fig, ax = plt.subplots(figsize=(12,7))
    lns1 = ax.plot(test['PRICE'],test['QUANTITY'], label='PRICE vs QUANTITY', color='orange')
    ax.set_xlabel('PRICE')
    ax.set_ylabel('QUANTITY')
    ax2=ax.twinx()
    
    lns2 = ax2.plot(test['PRICE'],test['PROFIT'], label='PRICE vs PROFIT', color='blue')
    ax2.set_ylabel('PROFIT')
    
    # added these three lines
    lns = lns1+lns2
    labs = [l.get_label() for l in lns]
    ax.legend(lns, labs, loc=0)
    
    ind = np.where(test['PROFIT'] == test['PROFIT'].max())[0][0]
    values_at_max_profit = test.iloc[[ind]]
    
    _ = ax.vlines(x=values_at_max_profit['PRICE'].values[0], ymin=0, 
                  ymax=values_at_max_profit['PROFIT'].values[0], colors='r', linestyle='--')
    
    
    
    plt.show()
    
    print(f"Getting MAX PROFIT {round(values_at_max_profit['PROFIT'].values[0],2)} at PRICE {round(values_at_max_profit['PRICE'].values[0],2)} for SKU {sku_name}")
    return values_at_max_profit
```
